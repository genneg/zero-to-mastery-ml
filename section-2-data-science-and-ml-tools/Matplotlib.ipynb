```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x=[1,2,3,4]
y=[11,22,33,44]
plt.plot(x,y)
plt.show()
```

```python
#1st method pyplot
fig= plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
#2nd method
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
```

```python
#3rd method
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x,y);
ax.set(title='Simple Plot', xlabel='x-axis', ylabel='y-axis')
fig.savefig('../images/sample-plot.png')
```
