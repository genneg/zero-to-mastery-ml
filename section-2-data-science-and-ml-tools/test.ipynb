```python
nut_butter = {'Almond':300, "Peanut":90, 'Jelly':120}
fig,ax = plt.subplots()
ax.set(title='Butter Store', xlabel='Butter type', ylabel='Sales')
ax.barh(list(nut_butter.keys()), list(nut_butter.values()));
```
